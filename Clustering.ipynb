## **K-Means dengan dataset berupa array**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Contoh data
data = np.array([[1, 2, 3, 4], [5, 8, 11, 14], [1.5, 1.8, 2.1, 2.4], [8, 8, 8, 8], [1, 0.6, 0.2, -0.2], [9, 11, 13, 15]])

# Jumlah kluster
k = 2

# Inisialisasi model K-Means
kmeans = KMeans(n_clusters=k)

# Melatih model
kmeans.fit(data)

centroids = kmeans.cluster_centers_
# Prediksi kluster untuk setiap data poin
labels = kmeans.predict(data)

print("Pusat Kluster:")
print(centroids)
print("Label Kluster:")
print(labels)

# Menghitung Silhouette Score
silhouette_avg = silhouette_score(data, labels)

print("Silhouette Score:", silhouette_avg)
```

## **Iterasi pada setiap jumlah kluster**
serta perhitungan skor siluet masing2 kluster

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

data_kucing = np.array([[1, 2, 3, 4], [5, 8, 11, 14], [1.5, 1.8, 2.1, 2.4], [8, 8, 8, 8], [1, 0.6, 0.2, -0.2], [9, 11, 13, 15]])
# Menentukan jumlah kluster
# X adalah data Anda
# Iterasi melalui berbagai nilai K
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_kucing)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    silhouette_avg = silhouette_score(data_kucing, labels)
    print("Pusat Kluster:")
    print(centroids)
    print("Label Kluster:")
    print(labels)
    print(f"For K={k}, Silhouette Score: {silhouette_avg}")
```

# **K-Means Clustering pada data kucing**
serta penentuan jumlah kluster yang tepat

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Data contoh untuk atribut jenis kelamin, umur, warna, dan status vaksinasi
data = {
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'age': [2, 1, 3, 2, 4, 3, 1, 2],
    'color': ['White', 'Black', 'Gray', 'White', 'Black', 'Gray', 'White', 'Black'],
    'vaccination_status': ['Vaccinated', 'Not Vaccinated', 'Vaccinated', 'Not Vaccinated',
                            'Vaccinated', 'Not Vaccinated', 'Vaccinated', 'Not Vaccinated']
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Select all features for clustering
features = ['gender', 'age', 'color', 'vaccination_status']
X = df[features]

# Convert categorical features to numerical using one-hot encoding
X = pd.get_dummies(X)

# Try different values of K and compute silhouette scores
silhouette_scores = []
possible_k_values = range(2, 6)  # You can adjust the range as needed

for k in possible_k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For K={k}, Silhouette Score: {silhouette_avg}")

# Plot the silhouette scores for different K values
plt.plot(possible_k_values, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal K')
plt.show()

# Choose the K with the highest silhouette score
optimal_k = possible_k_values[np.argmax(silhouette_scores)]
print(f"Optimal Number of Clusters (K): {optimal_k}")

# Fit the final K-means model with the optimal K
final_kmeans = KMeans(n_clusters=optimal_k, random_state=42)
final_cluster_labels = final_kmeans.fit_predict(X)

# Add the cluster labels back to your original DataFrame
df['cluster_label'] = final_cluster_labels

# Now 'df' DataFrame contains the cluster labels in the 'cluster_label' column
print(df)
```

# **Sistem Rekomendasi Bedasarkan Input User**

```python
# Function to get recommendations based on user input
def get_recommendations(user_input, df, kmeans_model):
    # Preprocess user input
    user_input_df = pd.DataFrame(user_input, index=[0])  # Create a DataFrame from user input

    # One-hot encode categorical features
    user_input_encoded = pd.get_dummies(user_input_df)

    # Add missing columns to user input DataFrame with default values
    missing_cols = set(X.columns) - set(user_input_encoded.columns)
    for col in missing_cols:
        user_input_encoded[col] = 0

    # Reorder columns to match the original DataFrame
    user_input_encoded = user_input_encoded[X.columns]

    # Predict the cluster for the user input
    user_cluster = kmeans_model.predict(user_input_encoded)[0]

    # Filter DataFrame for the same cluster
    cluster_df = df[df['cluster_label'] == user_cluster]

    # Exclude the user input from recommendations
    cluster_df = cluster_df[~cluster_df.index.isin(user_input_df.index)]

    return cluster_df

# Example user input
user_input = {
    'gender_Female': 1, #male = 0 female = 1
    'age': 2, #sesuai umur kucing
    'color_Black': 1, # white = 0 black =1 gray=2
    'vaccination_status_Not Vaccinated': 1 #Vaccinated =0, Not Vaccinated =1
}

# Get recommendations based on user input
recommendations = get_recommendations(user_input, df, final_kmeans)

# Display recommendations
print("Recommendations:")
print(recommendations)
```
